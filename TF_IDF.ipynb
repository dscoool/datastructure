```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
import json
```

```python
# ëŒ€ì²´í•œ corpus
# ìˆ˜ë™ìœ¼ë¡œ ì…ë ¥í•œ corpus ë‚´ìš©
corpus = [
'ê·¸ë£¹ ë°©íƒ„ì†Œë…„ë‹¨(BTS) ì •êµ­ì˜ ë‘ ë²ˆì§¸ ì†”ë¡œ ì‹±ê¸€ 3Dê°€ ì„¸ê³„ ì–‘ëŒ€ ì°¨íŠ¸ë¡œ ë¶ˆë¦¬ëŠ” ì˜êµ­ ì˜¤í”¼ì…œ ì‹±ê¸€ ì°¨íŠ¸ í†± 100ì— 3ì£¼ ì—°ì† ì§„ì…í–ˆë‹¤.20ì¼(í˜„ì§€ì‹œê°„) ê³µê°œëœ ìµœì‹  ì°¨íŠ¸ì— ë”°ë¥´ë©´ 3DëŠ” ì „ì£¼ë³´ë‹¤ ì—´ ê³„ë‹¨ í•˜ë½í•œ 32ìœ„ë¥¼ ê¸°ë¡í–ˆë‹¤.',
'3DëŠ” ë‹¿ì„ ìˆ˜ ì—†ëŠ” ìƒëŒ€ë°©ì— ëŒ€í•œ ë§ˆìŒì„ 1Â·2Â·3ì°¨ì›ì˜ ì‹œì„ ì´ë€ ì†Œì¬ë¡œ ì¬ë¯¸ìˆê²Œ í’€ì–´ë‚¸ ì•Œì•¤ë¹„(R&B) íŒ ì¥ë¥´ ë…¸ë˜ë‹¤.',
'ì´ ë…¸ë˜ëŠ” ì˜êµ­ ì‹±ê¸€ ì°¨íŠ¸ 5ìœ„ë¡œ ì²˜ìŒ ì§„ì…í•´ 22ìœ„, 32ìœ„ë¥¼ ê¸°ë¡í•˜ë©° ì¸ê¸°ë¥¼ ì´ì–´ê°€ê³  ìˆë‹¤.',
'ì„¸ê³„ë¥¼ ë¬´ëŒ€ë¡œ í™œë™í•˜ëŠ” í•œêµ­ì¸ DJ ê²¸ í”„ë¡œë“€ì„œ í˜ê¸° êµ¬ì˜ ë‚˜ë‚˜ë‚˜(NANANA)ëŠ” 42ìœ„ë¡œ 18ì£¼ ì—°ì† ì§„ì…í•´ ë¡±ëŸ°ì„ ì´ì–´ê°”ë‹¤.',
'í•œí¸, ì •êµ­ì˜ ì†”ë¡œ ë°ë·”ê³¡ ì„¸ë¸(Seven)ì€ ì„¸ê³„ ìµœëŒ€ ìŠ¤íŠ¸ë¦¬ë° í”Œë«í¼ ìŠ¤í¬í‹°íŒŒì´ê°€ ë°œí‘œí•œ ìœ„í´ë¦¬ í†± ì†¡ ê¸€ë¡œë²Œì—ì„œ 2ìœ„ë¡œ 14ì£¼ ì—°ì† ì§„ì…í–ˆë‹¤. 3DëŠ” 10ìœ„ë¡œ 3ì£¼ ì—°ì† í†± 10ì„ ì§€ì¼°ë‹¤.'
]
```

ë‰´ìŠ¤ì›ë³¸: https://news.kbs.co.kr/news/pc/view/view.do?ncd=7798660


```python
#TF-IDF
vect = CountVectorizer()                                #CountVectorizerë¥¼ ì´ìš©í•˜ì—¬
document_term_matrix = vect.fit_transform(corpus)
```

```python
# document_term_matrixë¥¼ dataframeìœ¼ë¡œ ë³€í™˜í•˜ê¸°!!
tf = pd.DataFrame(document_term_matrix.toarray(), columns=vect.get_feature_names_out())
# ë³€í™˜í•œ ë‚´ìš© ì¶œë ¥í•˜ê¸°!!
tf
```

```python
D = len(tf)
df = tf.astype(bool).sum(axis=0)
idf = np.log((D+1) / (df+1)) + 1
```

```python
df
```

```python
idf
```

```python
# TF-IDF (Term Frequency-Inverse Document Frequency)
tfidf = tf * idf
tfidf = tfidf / np.linalg.norm(tfidf, axis=1, keepdims=True) #ì •ê·œí™” normalisation
```

```python
tfidf
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
import json

#broadcast.json íŒŒì¼ì„ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤!!
# jsonì€ ë¹„ì •í˜•ë°ì´í„°(NoSQL)ì˜ ì¼ì¢…ì…ë‹ˆë‹¤. ê°•ì˜ê²Œì‹œíŒ ë‹¤ìš´ë¡œë“œ!!
with open('broadcast.json', encoding='UTF-8') as json_file:
    json_data = json.load(json_file)

# json_data
broadcasting_script = json_data["data"]
# jsonì˜ ë‚´ìš©ì„ ì¶œë ¥í•´ ë´…ì‹œë‹¤!!
broadcasting_script
```

ìë£Œê°€ ê¸¸ê¸° ë•Œë¬¸ì—, 100í–‰ê¹Œì§€ë§Œ ê°€ì ¸ì™€ ë³´ë„ë¡ í•˜ê² ìŠµë‹ˆë‹¤!!

forë¬¸ì„ ì‚¬ìš©í•˜ì—¬ [n]['ko']ë¥¼ ëª¨ë‘ ë¶ˆëŸ¬ì™€ì„œ </br>
'corpus'ë¼ëŠ” listì— ì €ì¥í•´ ë´…ì‹œë‹¤!!

corpus ë¼ëŠ” ë§ë­‰ì¹˜ë¥¼ ë§Œë“¤ì–´ ë´…ì‹œë‹¤. broadcasting_scriptì—ì„œ </br>
broadcasting_script[n]['ko']ì˜ ë‚´ìš©ë§Œ ëª¨ë‘ ì¶”ì¶œí•©ë‹ˆë‹¤!!

```python
corpus2 = []
for n, cor in enumerate(bs):
    corpus2.append(cor['ko'])
corpus2
```

ì´ì œ CounterVectorizer() ë¥¼ ì‚¬ìš©í•´ì„œ, </br>
ë¶ˆëŸ¬ì˜¨ ë¬¸ì¥ì— TF-IDFë¥¼ ì ìš©í•´ ë´…ì‹œë‹¤!!

#### TF (Term Frequency) ë¼ëŠ” ë§ì€ ë¬¸ì„œì—ì„œ </br>
#### ì–´ë– í•œ ë‹¨ì–´(Term)ì´ ëª‡ ë²ˆì´ë‚˜, ì–¼ë§ˆë‚˜ ìì£¼ ë‚˜ì˜¤ëŠ”ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì§€ìˆ˜ì…ë‹ˆë‹¤.</br>
ì˜ˆë¥¼ ë“¤ì–´, ë¬¸ì„œ 1ë²ˆì—ì„œ 'ì‚¬ê³¼'ë¼ëŠ” ë‹¨ì–´ê°€ 3ë²ˆ ë“±ì¥í•œë‹¤ë©´, </br>
DOC1 - 'ì‚¬ê³¼' ì˜ TFëŠ” 3ì…ë‹ˆë‹¤.</br>
í•˜ì§€ë§Œ í¸ì˜ë¥¼ ìœ„í•´ 3ìœ¼ë¡œ í‘œê¸°í•˜ì§€ ì•Šê³  í‘œì¤€í™”ë¥¼ ì‹œì¼œ </br>
(Normalization)í•˜ì—¬ í‘œê¸°í•©ë‹ˆë‹¤.


```python
#TF-IDF
vect = CountVectorizer()                                #CountVectorizerë¥¼ ì´ìš©í•˜ì—¬
document_term_matrix = vect.fit_transform(corpus2)       # ë¬¸ì„œ-ë‹¨ì–´ í–‰ë ¬ìƒì„±!!
```

TF-IDFë¥¼ ì ìš©í–ˆìŠµë‹ˆë‹¤. </br>
ì—¬ê¸°ì„œ ìƒì„±ëœ document_term_matrixë¥¼ </br>
1. tf ë¼ëŠ” ë°ì´í„°í”„ë ˆì„(DataFrame)ìœ¼ë¡œ ë³€í™˜í•˜ì—¬
2. ì¶œë ¥í•©ë‹ˆë‹¤!!!!

```python
# document_term_matrixë¥¼ dataframeìœ¼ë¡œ ë³€í™˜í•˜ê¸°!!
tf = pd.DataFrame(document_term_matrix.toarray(), columns=vect.get_feature_names())
# ë³€í™˜í•œ ë‚´ìš© ì¶œë ¥í•˜ê¸°!!
tf
```

```python
D = len(tf)
df = tf.astype(bool).sum(axis=0)
idf = np.log((D+1) / (df+1)) + 1             # IDF (Inverse Document Frequency)
```

```python
# TF-IDF (Term Frequency-Inverse Document Frequency)
tfidf = tf * idf
tfidf = tfidf / np.linalg.norm(tfidf, axis=1, keepdims=True)
```

```python
tfidf
```
